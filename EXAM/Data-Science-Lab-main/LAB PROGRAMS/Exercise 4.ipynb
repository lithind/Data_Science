```python
#simple calculator

a=int(input("Enter 1st number:"))
b=int(input("Enter 2nd number:"))
print("\n1.Addition")
print("2.Substraction")
print("3.Multipluication")
print("4.Division")
option=int(input("\nChoose an option:"))
if(option==1):
    print("Addition:",a+b)
elif(option==2):
    print("Substraction:",a-b)
elif(option==3):
    print("Multiplication",a*b)
elif(option==4):
    print("Division",a/b)
else:
    print("Invalid option")
```

```python
#dotted graph

from matplotlib import pyplot as pt
x=[1,2,3,4]
y=[1,2,5,7]
pt.plot(x,y,linewidth=2.0,linestyle="dashed")
pt.show()
```

```python
#program using grid

pt.grid()
pt.plot(x,y)
pt.show()
```

```python
#bar graph

from matplotlib import pyplot
pyplot.bar([1,2,3],[1,3,2])
pyplot.show()
```

```python
#pie chart

from matplotlib import pyplot
pyplot.pie([1,2,3],labels =['light','rain','forest'])
pyplot.show()
```

```python
#pie chart

from matplotlib import pyplot
x=[1,0,0]
pyplot.pie([1,2,3],labels =['light','rain','forest'],explode=x,shadow=True)
pyplot.show()
```

```python
#Histogram

import numpy as np
from matplotlib import pyplot as plt
x=np.random.normal(100,2,250)
plt.hist(x)
plt.show()
```
