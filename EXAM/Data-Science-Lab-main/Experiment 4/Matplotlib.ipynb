```python
from matplotlib import pyplot as pt
y=[5,4,3,2,1]
pt.plot(y)
pt.show()
```

```python
#plot graph

from matplotlib import pyplot as plt
a=[10,6,2,4]
plt.plot(a)
plt.show()
```

```python
#dotted graph

from matplotlib import pyplot as pt
x=[1,2,3,4]
y=[1,2,5,7]
pt.plot(x,y,linewidth=2.0,linestyle="dashed")
pt.show()
```

```python
#program using grid

pt.grid()
pt.plot(x,y)
pt.show()
```

```python
#plot graph with x and y rray values

import numpy as np
from matplotlib import pyplot as pt
x=np.arange(0,10,0.1)
y=np.sin(x)
pt.plot(x,y)
pt.xlabel("x axis")
pt.ylabel("y axis")
pt.title("sin graph")
pt.show()
```

```python
#to represent graph in scatter format

pt.scatter(x,y)
pt.show()
```

```python
#to color
pt.plot(x,y,'b+')
pt.show()
```

```python
from matplotlib import pyplot as pt
x=[0,1,2,3,4,5,6,7]
y=[1,5,7,9,11,13,15,17]
pt.plot(x,'r')
pt.plot(y,'g*')
pt.xlabel("x axis")
pt.ylabel("y axis")
pt.title("Graph")
pt.legend(['x','y'])
pt.show()
```

```python
#bar graph

from matplotlib import pyplot
pyplot.bar([1,2,3],[1,3,2])
pyplot.show()
```

```python
#pie chart

from matplotlib import pyplot
pyplot.pie([1,2,3],labels =['light','rain','forest'])
pyplot.show()
```

```python
#pie chart

from matplotlib import pyplot
x=[1,0,0]
pyplot.pie([1,2,3],labels =['light','rain','forest'],explode=x,shadow=True)
pyplot.show()
```

```python
#Histogram

import numpy as np
from matplotlib import pyplot as plt
x=np.random.normal(100,2,250)
plt.hist(x)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as pt
x=np.array([2,4,6,8])
y=np.array([2,4,6,8])
pt.barh(x,y,height=1.2)
pt.show()
```
