```python
#rotation

from scipy import ndimage
from matplotlib import pyplot as pt
import numpy as np
from PIL import Image
%matplotlib inline
a=Image.open('477393.jpg')
b=np.array(a)
c=ndimage.rotate(b,150)
d=Image.fromarray(c)
pt.imshow(d)
```

```python
#Scaling

from scipy import ndimage
from matplotlib import pyplot as pt
import numpy as np
from PIL import Image
%matplotlib inline
a=Image.open('477393.jpg')
b=np.array(a)
c=ndimage.zoom(b,(0.25,1,1))
d=Image.fromarray(c)
pt.imshow(d)
```

```python
#Shearing

import numpy as np
from scipy import ndimage
from matplotlib import pyplot as pt
import numpy as np
from PIL import Image
%matplotlib inline
a=Image.open('477393.jpg')
b=np.array(a)
h,w,s=b.shape
m=[[1,0,0],[0,1,0],[0,0,1]]
c=ndimage.affine_transform(b,m,offset=(0,h//2,0),output_shape=(h,h+w//2,s))
d=Image.fromarray(c)
pt.imshow(d)
```
